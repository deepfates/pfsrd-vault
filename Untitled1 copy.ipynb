## Extract the data (and metadata)

```python
from pathlib import Path
```

```python
!mkdir ../output
```

```python
out_dir = Path('./output')
input_dir = Path("raw/")
```

```python
import pandas as pd
```

```python
files = [pd.read_csv(f).fillna("?") for f in input_dir.glob('*.csv')]
```

```python
first_column_values = [f.iloc[:, 0].values for f in files]
```

```python
from itertools import chain

items = list(chain.from_iterable(first_column_values))
```

```python
import re
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext
```

```python
def clean_k(k):
    if "Unnamed" in k:
        return "Name"
    else:
        return k

def clean_v(v):
    if v == "?":
        return "0"
    else:
        return v

def tag_entities(entry: str, entities: list):
    text = "\n".join(f"{clean_k(k)}: {clean_v(v)}" for k, v in entry.items())
    cleaned_text = cleanhtml(text)
    for e in entities:
            if str(e) in text:
                labeled_e = f"[[{str(e)}]]"
                if labeled_e not in text:
                    text = text.replace(str(e), labeled_e)
    return(text)

# Get one row from the first file
first_row = files[0].iloc[0]
tag_entities(first_row, items)
```

```python
for entry in files:
        for row in entry.iterrows():
            print(tag_entities(row[1], items))
```
